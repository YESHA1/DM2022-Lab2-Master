```python
from collections import Counter
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from string import punctuation
import os
from tqdm import tqdm
import re
import numpy as np
import time
import pickle
from spacy.lang.en.stop_words import STOP_WORDS
```

```python
from imblearn.over_sampling import SMOTE
```

```python
ans_train['emotion'].value_counts()
```

```python
labels_ = []
```

```python
labels=ans_train[['emotion']]
labels=labels.values.tolist()
```

```python
texts=ans_train[['text']]
texts=texts.values.tolist()
```

```python
i=0
for label in labels:
    if label[0]=='joy':
        labels_.append(1)
    elif label[0]=='anticipation':
        labels_.append(2)
    elif label[0]=='trust':
        labels_.append(3)
    elif label[0]=='sadness':
        labels_.append(4)
    elif label[0]=='disgust':
        labels_.append(5)
    elif label[0]=='fear':
        labels_.append(6)
    elif label[0]=='anger':
        labels_.append(7)
    elif label[0]=='surprise':
        labels_.append(8)
    else:
        i+=1
labels=np.array(labels_)
del labels_
```

```python
len(labels)
```

```python
X_train, X_test, y_train, y_test = train_test_split(texts,labels,test_size=0.2,random_state=30)
```

```python
counter=Counter() #紀錄目前有哪些詞及他們的頻率
```

```python
#好方法
datas=[]
for text in texts:
    
    #print(text)
    
    text = "".join([c for c in text[0].lower() if c not in punctuation]) #去除特殊詞
    tmp = [item for item in text.split(" ") if len(item)>0 and item!='lh' and item not in STOP_WORDS]
    
    counter.update(tmp)
    #print(tmp)
    tmp=" ".join(tmp)
    
    datas.append(tmp)
    
    
    #print(text)
    #print(tmp)
```

```python
X_train, X_test, y_train, y_test = train_test_split(datas,labels,test_size=0.2,random_state=30)
```

```python
len(X_train)
```

```python
for sent in X_train :
    #print(sent)
    tmp = [item for item in sent.split(" ") if len(item)>0 and item!='lh' and item not in STOP_WORDS]
    #print(tmp)
    counter.update(tmp)
```

```python
print(X_train[10:20])
```

```python
print(np.max(list(counter.values())))
print(np.min(list(counter.values())))
print(len(counter))
```

```python
mean_value = np.mean(list(counter.values()))
min_value = mean_value/2  # mean_value/2
max_value = mean_value*8  # mean_value*8
counter = {k: v for k, v in counter.items() if v < max_value and v > min_value}
vocabulary = sorted(list(counter.keys()),reverse=True)

mean_value
```

```python
len(counter)
```

```python
len(vocabulary)
```

```python
tf = TfidfVectorizer(vocabulary=vocabulary)
```

```python
X_train=tf.fit_transform(X_train)
X_test=tf.transform(X_test)
```

```python
len(y_train)
```

```python
from imblearn.over_sampling import ADASYN
ADA=ADASYN(random_state=42)
```

```python
bX_train, by_train =ADA.fit_resample(X_train, y_train)
```

```python
pickle.dump({'X_train':X_train,'X_test':X_test,'y_train':y_train,'y_test':y_test},open('data_1500Feature_15W.pkl','wb'))
```

```python
#pickle.dump({'X_train':X_train,'X_test':X_test,'y_train':y_train,'y_test':y_test},open('data_1500Feature_15W.pkl','wb'))
```

```python
lr = LogisticRegression(max_iter=1000)
lr.fit(bX_train,by_train)
print('score:%.5f'%lr.score(X_test,y_test))
```
